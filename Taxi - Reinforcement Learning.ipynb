```python
import gym
```

We'll be using the Taxi-v2 environment for starters

```python
env = gym.make("Taxi-v2").env
```

```python
env.render()
```

The goal of the agent would be to pickup up the passenger from any of the 4 locations (R,G,B,Y)
The actions available are:
1) Up
2) Down
3) Left
4) Right
5) Pickup
6) Drop

```python
env.P
```

Movement -> -1, Pickup/DropOff -> -10, Done -> 20

```python
import numpy
```

```python
qTable = numpy.zeros([env.observation_space.n, env.action_space.n])
```

The Q Table is of 500x6 dimension: 6 action spaces for every 500 states

```python
import random

alpha = 0.1
discount_factor = 0.7
epsilon = 0.1

episodes = 10000

done = False

for i in range(episodes):
    state = env.reset()
    steps, reward = 0,0
    penalties = 0
    while not done:
        if(random.uniform(0,1) < epsilon):
            action = env.action_space.sample()
        else:
            action = numpy.argmax(qTable[state])

        next_s, reward, done, info = env.step(action)

        old_value = qTable[state, action]
        best_next = numpy.argmax(qTable[state])
        
        new_value = (1-alpha) * old_value + alpha * (reward + discount_factor * best_next)
        
        qTable[state, action] = new_value
        
        if reward == -10: 
            penalties += 1
            
        state = next_s
        steps += 1
        
    if(i%50 == 0):
        print("{} Episode done".format(i))
        
print("Training Finished")
```
